```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

X = np.arange(1, 50, 5)
```

```python
X
```

```python
len(X)
```

```python
y = X + 10
```

```python
y
```

```python
len(y)
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
```

```python
linear_regression_model = tf.keras.Sequential()
```

```python
linear_regression_model.add(tf.keras.layers.Dense(10, input_shape=[1], name='input_layer'))
linear_regression_model.add(tf.keras.layers.Dense(1, name='output_layer'))
```

```python
linear_regression_model.compile(loss='mae', optimizer='sgd', metrics=['mae'])
```

```python
linear_regression_model.fit(X, y, epochs=10)
```

```python
linear_regression_model.predict([9])
```

```python
linear_regression_model = tf.keras.Sequential()

linear_regression_model.add(tf.keras.layers.Dense(20, input_shape=[1], activation='relu', name='input_layer'))
linear_regression_model.add(tf.keras.layers.Dense(1, activation='linear', name='output_layer'))

adam = tf.keras.optimizers.Adam(learning_rate=0.01)

linear_regression_model.compile(loss='mae', optimizer=adam, metrics=['mae'])
```

```python
linear_regression_model.fit(X, y, epochs=300)
```

```python
linear_regression_model.predict([23])
```
